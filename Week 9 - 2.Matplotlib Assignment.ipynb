### Q1: What is Matplotlib? Why is it used? Name five plots that can be plotted using the Pyplot module of Matplotlib.
### Ans:-
Matplotlib is a widely used plotting library in Python that provides a variety of tools for creating static, interactive, and animated visualizations. It is particularly popular for its versatility and ease of use, making it a go-to choice for data visualization in scientific computing, engineering, and data analysis.

##### Why is it used?

1) Versatility: Matplotlib supports a wide range of plot types and customization options.

2) Integration: It integrates well with other Python libraries, like NumPy and pandas, facilitating complex data analysis.

3) Customization: It allows fine-grained control over plot appearance, including colors, fonts, and layout.

4) Interactivity: It supports interactive plots that can be embedded in applications or web frameworks.

5) Publication-Quality Graphics: It can generate high-quality plots suitable for publication and presentation.


##### Five plots that can be created using the Pyplot module of Matplotlib:

1) Line Plot: Useful for visualizing data trends over time or continuous variables. Example function: plt.plot().

2) Scatter Plot: Ideal for showing the relationship between two variables and identifying patterns or correlations. Example function: plt.scatter().

3) Bar Plot: Useful for comparing discrete quantities across different categories. Example function: plt.bar().

4) Histogram: Displays the distribution of a dataset by grouping data into bins. Example function: plt.hist().

5) Pie Chart: Useful for showing proportions and percentages of a whole. Example function: plt.pie().

### Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data plot a scatter plot.

import numpy as np

np.random.seed(3)

x = 3 + np.random.normal(0, 2, 50)

y = 3 + np.random.normal(0, 2, len(x))

Note: Also add title, xlabel, and ylabel to the plot.

### Ans:- 
A scatter plot is a type of data visualization that uses dots to represent the values of two different variables. Each dot on the plot corresponds to one data point, with its position determined by the values of the two variables. Scatter plots are useful for identifying relationships, correlations, and patterns between the variables.

Here's how you can generate and plot a scatter plot using the provided data and code:

```python
import numpy as np
import matplotlib.pyplot as plt

# Seed for reproducibility
np.random.seed(3)

# Generate data
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))

# Create a scatter plot
plt.scatter(x, y)

# Add title and labels
plt.title('Scatter Plot of Generated Data')
plt.xlabel('X values')
plt.ylabel('Y values')

# Show the plot
plt.show()
```

### Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.

Use the following data:

import numpy as np

For line 1: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([0, 100, 200, 300, 400, 500])

For line 2: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([50, 20, 40, 20, 60, 70])

For line 3: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([10, 20, 30, 40, 50, 60])

For line 4: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([200, 350, 250, 550, 450, 150])

### Ans:-
The subplot() function in Matplotlib is used to create multiple plots in a single figure. It allows you to specify the layout of the plots and which subplot each plot should be drawn in. This is useful for comparing multiple visualizations side-by-side or in a grid format.

To draw four line plots using subplot(), you can follow these steps:

1) Create a 2x2 grid of subplots. This will give you four areas within the figure where each plot will be drawn.
2) Plot each line in its designated subplot.

```python
# Exampleimport numpy as np
import matplotlib.pyplot as plt

# Define data for each line
x1 = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])

x2 = np.array([0, 1, 2, 3, 4, 5])
y2 = np.array([50, 20, 40, 20, 60, 70])

x3 = np.array([0, 1, 2, 3, 4, 5])
y3 = np.array([10, 20, 30, 40, 50, 60])

x4 = np.array([0, 1, 2, 3, 4, 5])
y4 = np.array([200, 350, 250, 550, 450, 150])

# Create a figure and a grid of subplots
fig, axs = plt.subplots(2, 2, figsize=(10, 8))

# Plot each line in a different subplot
axs[0, 0].plot(x1, y1, 'r-o')  # Line 1 in subplot (0, 0)
axs[0, 0].set_title('Line 1')
axs[0, 0].set_xlabel('X values')
axs[0, 0].set_ylabel('Y values')

axs[0, 1].plot(x2, y2, 'g--s')  # Line 2 in subplot (0, 1)
axs[0, 1].set_title('Line 2')
axs[0, 1].set_xlabel('X values')
axs[0, 1].set_ylabel('Y values')

axs[1, 0].plot(x3, y3, 'b-^')  # Line 3 in subplot (1, 0)
axs[1, 0].set_title('Line 3')
axs[1, 0].set_xlabel('X values')
axs[1, 0].set_ylabel('Y values')

axs[1, 1].plot(x4, y4, 'm-x')  # Line 4 in subplot (1, 1)
axs[1, 1].set_title('Line 4')
axs[1, 1].set_xlabel('X values')
axs[1, 1].set_ylabel('Y values')

# Adjust layout to prevent overlap
plt.tight_layout()

# Show the plot
plt.show()
```

### Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.

import numpy as np

company = np.array(["Apple", "Microsoft", "Google", "AMD"])

profit = np.array([3000, 8000, 1000, 10000])
 

### Ans:-
A bar plot (or bar chart) is a type of data visualization that represents categorical data with rectangular bars. The length or height of each bar is proportional to the value it represents. Bar plots are useful for comparing different categories or groups side-by-side to easily see which category has the highest or lowest values.

Why is it used?

1) Comparison: Bar plots are excellent for comparing quantities across different categories.
2) Clarity: They clearly show the differences between categories, making them useful for presenting data in a straightforward manner.
3) Categorical Data: Ideal for visualizing data that is divided into categories or groups.

```python
import numpy as np
import matplotlib.pyplot as plt

# Define data
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

# Create a figure and axes
fig, axs = plt.subplots(1, 2, figsize=(14, 6))

# Vertical Bar Plot
axs[0].bar(company, profit, color='skyblue')
axs[0].set_title('Vertical Bar Plot')
axs[0].set_xlabel('Company')
axs[0].set_ylabel('Profit')

# Horizontal Bar Plot
axs[1].barh(company, profit, color='salmon')
axs[1].set_title('Horizontal Bar Plot')
axs[1].set_xlabel('Profit')
axs[1].set_ylabel('Company')

# Adjust layout to prevent overlap
plt.tight_layout()

# Show the plot
plt.show()
```

### Q5: What is a box plot? Why is it used? Using the following data plot a box plot.

box1 = np.random.normal(100, 10, 200)

box2 = np.random.normal(90, 20, 200)
### Ans:-
A box plot (or box-and-whisker plot) is a graphical representation of the distribution of a dataset. It provides a summary of key statistical measures, including the median, quartiles, and potential outliers.

Why is it used?

1) Distribution Overview: Box plots offer a clear summary of the distribution of data, including its central tendency and variability.
2) Identifying Outliers: They help identify outliers or extreme values that fall outside the typical range of the data.
3) Comparing Distributions: Multiple box plots can be compared side-by-side to assess differences between distributions, such as between different groups or conditions.
4) Understanding Spread: They provide insights into the spread and symmetry of the data.

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate data
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

# Create a figure and axis
fig, ax = plt.subplots(figsize=(8, 6))

# Plot the box plot
ax.boxplot([box1, box2], labels=['Box 1', 'Box 2'], patch_artist=True)

# Add title and labels
ax.set_title('Box Plot of Two Datasets')
ax.set_xlabel('Group')
ax.set_ylabel('Value')

# Show the plot
plt.show()
```
