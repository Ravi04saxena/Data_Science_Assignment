### Q1: What are the Probability Mass Function (PMF) and Probability Density Function (PDF)? Explain with an example.
### Sol:- 
#### 1. Probability Mass Function (PMF) :- The PMF is used for discrete random variables. It gives the probability of each possible outcome in a discrete probability distribution. A random variable is discrete if it takes on a countable number of distinct values (e.g., rolling a die, number of heads in coin tosses).
* Definition: The PMF of a discrete random variable X is a function that maps each possible outcome x to the probability that X takes that value.


#### 2. Probability Density Function (PDF) :- The PDF is used for continuous random variables. A random variable is continuous if it takes on an infinite number of possible values within a given range (e.g., height, weight, time).
* Definition: The PDF of a continuous random variable X is a function that describes the relative likelihood of X taking a particular value within a range. However, unlike discrete variables, the probability that X takes any exact value is 0. Instead, the probability is determined over an interval, and the PDF is integrated over that interval.|

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom, norm

# 1. PMF Example (Discrete) - Binomial Distribution
n = 10  # Number of trials
p = 0.5  # Probability of success

# Generate the PMF for values from 0 to n
x_binom = np.arange(0, n+1)
pmf_binom = binom.pmf(x_binom, n, p)

# Plotting the PMF
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.stem(x_binom, pmf_binom, basefmt=" ", use_line_collection=True)
plt.title('PMF of Binomial Distribution (n=10, p=0.5)')
plt.xlabel('Number of Successes (k)')
plt.ylabel('Probability')
plt.grid(True)

# 2. PDF Example (Continuous) - Normal Distribution
mu = 0  # Mean
sigma = 1  # Standard deviation

# Generate values from -4 to 4 for the PDF
x_norm = np.linspace(-4, 4, 1000)
pdf_norm = norm.pdf(x_norm, mu, sigma)

# Plotting the PDF
plt.subplot(1, 2, 2)
plt.plot(x_norm, pdf_norm, label=f'N({mu}, {sigma}^2)')
plt.title('PDF of Normal Distribution (mu=0, sigma=1)')
plt.xlabel('x')
plt.ylabel('Probability Density')
plt.grid(True)
plt.tight_layout()

# Show plots
plt.show()
```

### Q2: What is Cumulative Density Function (CDF)? Explain with an example. Why CDF is used?
#### Sol:- Cumulative Distribution Function (CDF) : The Cumulative Distribution Function (CDF) is used to describe the probability that a random variable X takes a value less than or equal to a given point. It gives the cumulative probability up to that value, providing the probability that the variable is less than or equal to that threshold.

* For Discrete Random Variables: The CDF is the sum of the probabilities of all outcomes less than or equal to a certain value.
* For Continuous Random Variables: The CDF is the integral of the probability density function (PDF) up to a given point.

##### Mathematical Definition:
For a random variable X, the CDF F(x) is defined as:
   * F(x)=P(Xx)

     * Where:
        * F(x) is the probability that X takes a value less than or equal to x.

#### Why CDF is Used?
* Cumulative Probability: The CDF gives the total probability up to a point, making it easier to understand how likely it is for a random variable to be below a certain value.
* Distribution Description: It describes the entire distribution of a random variable and can be used to compute probabilities for intervals by subtracting values of the CDF.
* Non-decreasing Function: The CDF is always non-decreasing, meaning that as x increases, the cumulative probability either increases or stays the same.
* Useful for Percentiles: The CDF helps find percentiles of distributions, allowing us to calculate thresholds for specific probabilities.

#### Example of CDF:
##### For Discrete Random Variable (Binomial Distribution):
* Lets say you have a binomial distribution where you flip a coin 3 times, and the probability of success (heads) is 0.5. The possible number of heads (successes) could be 0, 1, 2, or 3.

* The PMF (Probability Mass Function) tells us the probability of each outcome. The CDF at each point is the cumulative probability of getting that many or fewer successes.
  * P(X=0)=0.125
  * P(X=1)=0.375
  * P(X=2)=0.375
  * P(X=3)=0.125

* The CDF is calculated by summing up probabilities up to each outcome:
  * F(0)=P(X0)=0.125
  * F(1)=P(X1)=0.125+0.375=0.5
  * F(2)=P(X2)=0.5+0.375=0.875
  * F(3)=P(X3)=0.875+0.125=1

##### For Continuous Random Variable (Normal Distribution):
* For a normal distribution with mean =0 and standard deviation =1, the CDF is calculated by integrating the PDF from  to a value x.

* For example, the CDF at x=1 tells us the probability that a normally distributed random variable is less than or equal to 1. This would be the area under the normal curve from  to 1.

   * For a standard normal distribution, P(X1)0.8413, meaning theres an 84.13% chance that the value will be less than or equal to 1.


```python
# Example
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom, norm

# 1. CDF Example for a Binomial Distribution
n = 3  # Number of trials
p = 0.5  # Probability of success

# Generate the CDF for values from 0 to n
x_binom = np.arange(0, n+1)
cdf_binom = binom.cdf(x_binom, n, p)

# Plotting the CDF for Binomial Distribution
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.step(x_binom, cdf_binom, where='mid', label='CDF', color='blue')
plt.title('CDF of Binomial Distribution (n=3, p=0.5)')
plt.xlabel('Number of Successes (k)')
plt.ylabel('Cumulative Probability')
plt.grid(True)

# 2. CDF Example for a Normal Distribution
mu = 0  # Mean
sigma = 1  # Standard deviation

# Generate values from -4 to 4 for the CDF
x_norm = np.linspace(-4, 4, 1000)
cdf_norm = norm.cdf(x_norm, mu, sigma)

# Plotting the CDF for Normal Distribution
plt.subplot(1, 2, 2)
plt.plot(x_norm, cdf_norm, label=f'N({mu}, {sigma}^2)', color='green')
plt.title('CDF of Normal Distribution (mu=0, sigma=1)')
plt.xlabel('x')
plt.ylabel('Cumulative Probability')
plt.grid(True)

# Show plots
plt.tight_layout()
plt.show()
```

### Q3: What are some examples of situations where the normal distribution might be used as a model? Explain how the parameters of the normal distribution relate to the shape of the distribution.
### Sol :- Situations Where the Normal Distribution Might Be Used:
#### 1. Height of People: 
The distribution of adult human heights often follows a normal distribution, with most people falling around an average height (the mean), and fewer people being extremely short or tall.
#### 2. IQ Scores: 
Intelligence test scores such as IQ are often modeled using a normal distribution, where the majority of people score near the mean, and fewer people score very high or very low.
#### 3. Measurement Errors: 
In experiments, random measurement errors often follow a normal distribution, assuming the errors are unbiased and independent.
#### 4. Stock Market Returns: 
The daily returns of some financial assets can be approximated by a normal distribution, especially over short time intervals.

### Parameters of the Normal Distribution and How They Affect the Shape:
The normal distribution is defined by two parameters: mean () and standard deviation ().
#### 1. Mean ():
The mean determines the center of the distribution, or the location of the peak. A change in the mean shifts the distribution left or right without altering its shape.
#### 2. Standard Deviation ():
The standard deviation controls the spread or width of the distribution. A larger standard deviation results in a wider, flatter curve, indicating more variability in the data. Conversely, a smaller standard deviation leads to a narrower, taller curve, indicating that most data points are closer to the mean.
### Shape and Properties of the Normal Distribution:
* Symmetry: The normal distribution is symmetric around the mean.
* Bell Shape: The curve is bell-shaped, with the highest point at the mean, and it tapers off as you move away from the mean.
* 68-95-99.7 Rule: Approximately 68% of the data lies within one standard deviation of the mean, 95% within two, and 99.7% within three.

In summary, the mean shifts the normal distribution left or right, while the standard deviation controls how spread out or concentrated the data is around the mean.

### Q9: What is Central Limit Theorem? State the significance of the Central Limit Theorem.
### Sol:- Central Limit Theorem (CLT):
The Central Limit Theorem (CLT) is one of the most fundamental theorems in statistics. It states that the sampling distribution of the sample mean (or sum) of a sufficiently large number of independent, identically distributed (i.i.d.) random variables will be approximately normally distributed, regardless of the shape of the original population distribution, as long as the sample size is large enough.

#### Key Conditions:
* Large Sample Size: For the CLT to hold, the sample size n should generally be 30 or more. However, if the population distribution is normal, the CLT applies even for smaller sample sizes.
* Independence: The samples must be independent of each other.
* Identical Distribution: The population from which the samples are drawn must have the same distribution.

#### Significance of the Central Limit Theorem:
##### 1.Foundation for Inferential Statistics:
* The CLT forms the basis for many statistical methods, including confidence intervals and hypothesis tests. Even if the underlying population is not normally distributed, the CLT assures that the sample means will follow a normal distribution, allowing us to apply parametric tests like the z-test or t-test.

##### 2.Approximation of Distribution:
* The CLT allows us to approximate the distribution of the sample mean with a normal distribution, even when the original data does not follow a normal distribution. This is especially useful because normal distributions are well understood and easier to work with.

##### 3.Simplification of Complex Problems:
* Many real-world problems involve non-normal data. The CLT enables the use of normal distribution approximations for these problems, simplifying statistical analysis.

##### 4.Probability Calculations:
* Because of the CLT, we can compute probabilities about sample means using the standard normal distribution (z-scores). For example, if we want to know the probability that the average of a sample falls within a certain range, we can use the normal distribution to make this calculation.

##### 5.Large-Scale Applications:
* The CLT is widely used in fields such as finance (e.g., modeling stock prices), economics, quality control, and engineering, where aggregate outcomes of many random variables are often approximated by a normal distribution.

##### 6.Sampling Theory:
* The CLT justifies why taking random samples from a population is an effective way of estimating population parameters. As the sample size increases, the sample mean becomes a better estimate of the population mean, which leads to more accurate inferences.

### Q10: State the assumptions of the Central Limit Theorem.
### Sol:- The Central Limit Theorem (CLT) has several key assumptions that must be met for it to hold true and for the sampling distribution of the sample mean to approximate a normal distribution. These assumptions ensure that the CLT applies regardless of the original population's distribution.

#### Assumptions of the Central Limit Theorem:
##### 1.Random Sampling:
* The data must be obtained from a random sample. This means that each observation in the sample is randomly selected from the population, and every possible sample has an equal chance of being chosen. Random sampling helps ensure that the sample is representative of the population.

##### 2.Independence of Observations:
* The observations in the sample must be independent of each other. This means that the value of one observation should not influence the value of another. If observations are dependent, such as in clustered or time-series data, the CLT may not apply directly.
* Independence is typically ensured if the sampling is done with replacement. If sampling is done without replacement, independence is usually maintained as long as the sample size is small relative to the population size (i.e., no more than 10% of the population).

##### 3.Identical Distribution (i.i.d. Assumption):
* The observations should come from the same distribution (i.e., they should be identically distributed). This means that each data point is drawn from the same population, and the distribution of each data point is identical. If observations come from different distributions, the CLT may not hold.

##### 4.Sample Size:
* The sample size should be sufficiently large. In general, a sample size of 30 or more is considered large enough for the CLT to hold, especially if the population distribution is unknown or not normally distributed.
  * If the population is normally distributed, even a small sample size (e.g., n<30) is sufficient for the sample mean to be normally distributed.
  * For heavily skewed or highly non-normal populations, a larger sample size is needed for the CLT to apply.

#### 5.Finite Variance:
* The population from which the sample is drawn should have a finite variance (^2). If the variance is infinite, the sample mean may not converge to a normal distribution as the sample size increases.

### Summary of Assumptions:
* Random sampling: Data should come from a random process.
* Independence: Observations must be independent of each other.
* Identically distributed: Data should come from the same distribution.
* Sufficiently large sample size: Typically n30, but larger may be required for non-normal populations.
* Finite variance: The population should have a finite variance.

These assumptions ensure that the CLT can be applied in practice, leading to a normally distributed sample mean even when the population distribution is unknown or non-normal.
